Bài tập: sinh viên tìm hiểu các hàm trên qua vd

```python
import tensorflow as tf

# KHỞI TẠO CÁC TENSOR BAN ĐẦU
t1 = tf.constant([4., 3., 2.])
t2 = tf.constant([3., 2., 1.])
t3 = tf.constant([[1, 2, 3],
                  [4, 5, 6]])
t4 = tf.constant([[7, 8],
                  [9, 10],
                  [11, 12]])

# 1. TÍCH CO TENSOR (TENSOR DOT PRODUCT)
# axes=0: Nhân tensor outer product (tích ngoài), tạo ra ma trận kết hợp bậc cao hơn.
dot_x = tf.tensordot(t1, t2, 0)
# axes=1: Nhân vô hướng hai vector (Dot Product thông thường) -> Kết quả: 4*3 + 3*2 + 2*1 = 20.0
dot_y = tf.tensordot(t1, t2, 1)

# LỆNH IN THỬ KIỂM TRA DỮ LIỆU GỐC
print(t1)
print(t2)
print(t3)
print(t4)

# 2. TÍCH CÓ HƯỚNG GIỮA HAI VECTOR (CROSS PRODUCT)
# Chỉ áp dụng cho vector 3 chiều, trả về một vector mới vuông góc với mặt phẳng chứa t1 và t2.
cross = tf.linalg.cross(t1, t2)

# 3. PHÉP NHÂN MA TRẬN ĐẠI SỐ TUYẾN TÍNH (MATRIX MULTIPLICATION)
# Nhân hàng của t3 (2x3) với cột của t4 (3x2) để tạo ra một ma trận mới có kích thước (2x2).
mm = tf.matmul(t3, t4)

# 4. RÚT TRÍCH ĐƯỜNG CHÉO CHÍNH (DIAGONAL)
# Lấy ra các phần tử nằm trên đường chéo chính của ma trận t3.
d = tf.linalg.diag(t3)

# 5. TÍNH VẾT CỦA MA TRẬN (TRACE)
# Tổng của các phần tử nằm trên đường chéo chính của ma trận.
trace = tf.linalg.trace(t3)

# 6. PHÉP CHUYỂN VỊ MA TRẬN (TRANSPOSE)
# Biến các hàng của ma trận t3 thành các cột của ma trận mới (kích thước chuyển từ 2x3 sang 3x2).
trans = tf.transpose(t3)

# 7. KHỞI TẠO MA TRẬN ĐƠN VỊ (IDENTITY MATRIX)
# Tạo một ma trận kích thước 5x5 chứa toàn số 0, ngoại trừ đường chéo chính chứa toàn số 1.
eye = tf.eye(5, 5)

# ĐOẠN XUẤT KẾT QUẢ RA MÀN HÌNH THEO HÌNH ẢNH
print(dot_x)
print(d)
print(cross)
print(trace)
print(trans)
print(eye)
```

```python
import tensorflow as tf

# Định nghĩa một hàm Python thông thường (Chạy ở chế độ Eager Execution)
def a_regular_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x

# CHUYỂN ĐỔI SANG GRAPH: Bọc hàm Python thông thường lại thành một TensorFlow Function
# Lúc này, TensorFlow sẽ biên dịch tĩnh các phép toán bên trong thành một Đồ thị tính toán (Computation Graph)
a_function_that_uses_a_graph = tf.function(a_regular_function)

# Khởi tạo các Tensor kiểm thử
x1 = tf.constant([[1.0, 2.0]])
y1 = tf.constant([[2.0], [3.0]])
b1 = tf.constant(4.0)

# Gọi hàm chạy ở chế độ thông thường (Eager Mode)
orig_value = a_regular_function(x1, y1, b1).numpy()

# Gọi hàm đã được tối ưu hóa bằng Graph (Graph Mode)
tf_function_value = a_function_that_uses_a_graph(x1, y1, b1).numpy()

# Xuất kết quả nghiệm thu (Cả hai chế độ đều cho ra cùng một đáp án số học là 12.0)
print(orig_value)
print(tf_function_value)
```

```python
import tensorflow as tf

# Định nghĩa một hàm Python thông thường (Không có @tf.function)
def inner_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x

# Sử dụng decorator @tf.function để biến hàm 'outer_function' thành một TensorFlow Graph Function
@tf.function
def outer_function(x):
    y = tf.constant([[2.0], [3.0]])
    b = tf.constant(4.0)

    # Hàm outer gọi hàm inner bên trong nó
    return inner_function(x, y, b)

# Gọi thực thi hàm với một Tensor đầu vào
# Dòng chú thích trong hình: Note that the callable will create a graph that includes `inner_function` as well as `outer_function`.
rs = outer_function(tf.constant([[1.0, 2.0]])).numpy()
print(rs)
```

3.18. Modeles, layers và models

Ví dụ tạo lớp SimpleModule xử lý tensor scalar (Layer)

```python
import tensorflow as tf

# Tạo một lớp Module tùy biến bằng cách kế thừa tf.Module
class SimpleModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)

        # 1. Biến có thể huấn luyện (Trainable Variable - Trọng số học tập chuẩn)
        self.a_variable = tf.Variable(5.0, name="train_me")

        # 2. Biến đóng băng / cố định (Non-trainable Variable - Không tham gia huấn luyện)
        # Đặt thuộc tính trainable=False để thuật toán Gradient Descent bỏ qua không cập nhật biến này
        self.non_trainable_variable = tf.Variable(5.0, trainable=False, name="do_not_train_me")

    def __call__(self, x):
        # Định nghĩa phép toán Forward Pass (Lan truyền tiến): y = a * x + b
        return self.a_variable * x + self.non_trainable_variable

# Khởi tạo một đối tượng từ lớp vừa định nghĩa
simple_module = SimpleModule(name="simple")

# Gọi thực thi tính toán Forward với đầu vào x = 5.0 -> Kết quả: 5.0 * 5.0 + 5.0 = 30.0
rs = simple_module(tf.constant(5.0))
print(rs)

# TRÍCH XUẤT QUẢN LÝ BIẾN CỦA TENSORFLOW
print('Train variables', simple_module.trainable_variables)
print('All variables', simple_module.variables)
```

Ví dụ tạo model với nhiều tầng layer

```python
import tensorflow as tf

# 1. Định nghĩa tầng liên kết đầy đủ (Dense Layer / Fully Connected Layer)
class Dense(tf.Module):
    def __init__(self, in_features, out_features, name=None):
        super().__init__(name=name)
        # Khởi tạo ma trận trọng số w ngẫu nhiên theo phân phối chuẩn
        self.w = tf.Variable(
            tf.random.normal([in_features, out_features]), name='w')
        # Khởi tạo độ lệch b chứa toàn số 0
        self.b = tf.Variable(tf.zeros([out_features]), name='b')

    def __call__(self, x):
        # Tính Forward Pass: y = (x * w) + b
        y = tf.matmul(x, self.w) + self.b
        # Đi qua hàm kích hoạt ReLU để khử tuyến tính (Giữ lại số dương, ép số âm về 0)
        return tf.nn.relu(y)

# 2. Kết hợp các tầng đơn lẻ thành một chuỗi mạng nơ-ron (Sequential Model)
class SequentialModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        # Tầng ẩn 1: Nhận vào 3 đặc trưng, trích xuất ra 3 đặc trưng
        self.dense_1 = Dense(in_features=3, out_features=3)
        # Tầng đầu ra 2: Nhận vào 3 đặc trưng từ tầng 1, đưa ra 2 kết quả dự đoán cuối cùng
        self.dense_2 = Dense(in_features=3, out_features=2)

    def __call__(self, x):
        # Dữ liệu đi tuần tự qua tầng 1 rồi ném kết quả vào tầng 2
        x = self.dense_1(x)
        return self.dense_2(x)

# Khởi tạo mô hình
# Dòng chú thích: You have made a model!
my_model = SequentialModule(name="the_model")

# Gọi mô hình với dữ liệu đầu vào giả lập là một vector kích thước (1, 3)
# Dòng chú thích: Call it, with random results
print("Model results:", my_model(tf.constant([[2.0, 2.0, 2.0]])))
```

3.19.1: Lưu checkpoint

```python
import tensorflow as tf

# 1. Định nghĩa đường dẫn/tên file để lưu trữ checkpoint
chkp_path = "my_checkpoint"

# 2. Khởi tạo một đối tượng Checkpoint và gắn mô hình của cậu vào tham số quản lý
# (Ở đây 'my_model' chính là đối tượng mạng SequentialModule cậu đã tạo ở câu trước)
checkpoint = tf.train.Checkpoint(model=my_model)

# 3. Ghi/Lưu toàn bộ giá trị trọng số (w) và bias (b) hiện tại xuống đĩa cứng
checkpoint.write(chkp_path)
```

```python
import tensorflow as tf

# 1. Khởi tạo một đối tượng mô hình hoàn toàn mới (Trọng số lúc này đang là ngẫu nhiên)
new_model = SequentialModule()

# 2. Tạo một đối tượng Checkpoint mới và gắn mô hình mới vào để quản lý
new_checkpoint = tf.train.Checkpoint(model=new_model)

# 3. Nạp/Khôi phục lại toàn bộ trọng số cũ từ file "my_checkpoint" đè lên mô hình mới
new_checkpoint.restore("my_checkpoint")

# 4. Kiểm tra độ đồng nhất: Dự báo với cùng một đầu vào [[2.0, 2.0, 2.0]]
# Dòng chú thích: Should be the same result as above
print(new_model(tf.constant([[2.0, 2.0, 2.0]])))
```

```python
import tensorflow as tf

# Khởi tạo ma trận X ban đầu
X = tf.constant([[1, 2, 3],
                 [4, 5, 6]])

# Tạo tensor chứa toàn số 0 có cùng shape và dtype với X
X_zero = tf.zeros_like(X)

print(X_zero)
```

Cách 2: Sử dụng thuộc tính .shape và .dtype của X

```python
import tensorflow as tf

X = tf.constant([[1, 2, 3],
                 [4, 5, 6]])

# Lấy thủ công thông tin kích thước và kiểu dữ liệu của X truyền vào tf.zeros
X_zero_manual = tf.zeros(shape=X.shape, dtype=X.dtype)

print(X_zero_manual)
```

3. Tạo một tensor có kích thước [3, 2] với các phần tử được gán 5

Cách 1: Sử dụng hàm tf.fill

```python
import tensorflow as tf

# Tạo tensor kích thước [3, 2] chứa toàn số 5
t_five = tf.fill([3, 2], 5)

print(t_five)
```

Cách 2: Sử dụng toán tử nhân với tf.ones

```python
import tensorflow as tf

# Tạo ma trận toàn số 1 rồi nhân với 5
t_five_alt = tf.ones([3, 2], dtype=tf.int32) * 5

print(t_five_alt)
```

5. Tạo một tensor 1-chiều có giá trị như [10, 12, 14, …, 100]

```python
import tensorflow as tf

# Tạo tensor 1-chiều chạy từ 10 đến 100 với bước nhảy là 2
# Lưu ý: Điểm kết thúc (limit) phải đặt là 101 để chuỗi lấy được cả số 100
t_range = tf.range(start=10, limit=101, delta=2)

print(t_range)
```

6. Tạo một tensor kích thước [3, 2] có giá trị ngẫu nhiên theo phân phối normal với mean=0,
standard deviation = 2

```python
import tensorflow as tf

# Tạo tensor ngẫu nhiên phân phối chuẩn kích thước [3, 2]
# Định rõ mean = 0.0 và độ lệch chuẩn (stddev) = 2.0
t_random_normal = tf.random.normal(shape=[3, 2], mean=0.0, stddev=2.0)

print(t_random_normal)
```

7. Tạo một tensor kích thước [3, 2] có giá trị ngẫu nhiên theo phân phối uniform trong phạm vi
[0, 2]

```python
import tensorflow as tf

# Tạo tensor ngẫu nhiên phân phối đều kích thước [3, 2]
# Định rõ giá trị nhỏ nhất (minval) = 0.0 và lớn nhất (maxval) = 2.0
t_random_uniform = tf.random.uniform(shape=[3, 2], minval=0.0, maxval=2.0)

print(t_random_uniform)
```

8. Cho tensor X = [[1, 2], [3, 4], [5, 6]]. Hãy shuffle X theo chiều thứ nhất (sử dụng
random_shuffle)

```python
import tensorflow as tf

# Khởi tạo ma trận X ban đầu kích thước [3, 2]
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])

# Xáo trộn ngẫu nhiên X theo chiều thứ nhất (trục các hàng)
shuffled_X = tf.random.shuffle(X)

print(shuffled_X)
```

9. Hãy tạo ngẫu nhiên tensor kích thước (10, 10, 3). Cắt thành tensor có kích thước (5, 5, 3) (sử
dụng random_crop)

```python
import tensorflow as tf

# 1. Tạo ngẫu nhiên một tensor kích thước (10, 10, 3)
# Ở đây tớ dùng phân phối đều (Uniform) trong khoảng [0, 255] để giả lập một tấm ảnh màu RGB
original_tensor = tf.random.uniform(shape=[10, 10, 3], minval=0, maxval=255, dtype=tf.int32)

# 2. Thực hiện cắt ngẫu nhiên về kích thước (5, 5, 3)
# Tham số size nhận vào một mảng định hình cấu trúc kích thước mục tiêu muốn giữ lại
cropped_tensor = tf.image.random_crop(original_tensor, size=[5, 5, 3])

print("Kích thước tensor gốc:", original_tensor.shape)
print("Kích thước tensor sau khi cắt:", cropped_tensor.shape)
```
