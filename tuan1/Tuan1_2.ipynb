Thay đổi giá trị của ma trận (Sự khác biệt giữa sao chép và tham chiếu)

```python
# Tạo ma trận 5 x 5 toàn số 1
A = np.matrix(np.ones((5, 5)))
print('Ma trận gốc A:\n', A)

# Sử dụng np.array tạo bản sao mới (không ảnh hưởng ma trận gốc A)
B = np.array(A)
B[2] = 2
print('Thay đổi dòng 3 trên bản sao bằng np.array (A không đổi):\n', A)

# Sử dụng np.asarray tạo tham chiếu (thay đổi trực tiếp lên ma trận gốc A)
np.asarray(A)[2] = 2
print('Thay đổi dòng 3 bằng np.asarray (A bị thay đổi):\n', A)
```

Sinh dãy số theo khoảng cho trước (Arange & Linspace)

```python
# Hàm arange: khởi đầu, kết thúc (không gồm n), bước nhảy
a = np.arange(1, 20)
print("Arange bước nhảy 1:\n", a)

b = np.arange(1, 20, 2)
print("Arange bước nhảy 2:\n", b)

# Hàm linspace: khởi đầu, kết thúc, số lượng phần tử được chia đều
c = np.linspace(1., 5., num=10)
print("Linspace chia đều 10 phần tử từ 1 đến 5:\n", c)
```

3.10. Index và Slice (Cắt mảng và Truy xuất phần tử)

```python
a = np.array([(1, 2, 3), (4, 5, 6)])
print("Mảng gốc:\n", a)
print('Dòng đầu tiên (row 0):', a[0])
print('Dòng thứ hai (row 1):', a[1])

# Indexing & Slicing dòng/cột
print('Cột thứ hai (column 1):', a[:, 1])
print('Cột đầu tiên (column 0):', a[:, 0])
print('Phần tử tại dòng 1, cột 2 (số 6):', a[1, 2])
```

Thao tác nâng cao trên DataFrame (Nối, Lọc, Xóa cột) - Từ tài liệu hình ảnh

```python
# Tạo dữ liệu mẫu
df1 = pd.DataFrame({'Name': ['John', 'Smith', 'Paul'], 'Age': [25, 30, 50]}, index=[1, 2, 3])
df2 = pd.DataFrame({'Name': ['Adam', 'Smith'], 'Age': [28, 11]}, index=[4, 5])

# 7.1. Nối nhiều DataFrame
df_concat = pd.concat([df1, df2])
print("Sau khi nối DataFrame:\n", df_concat)

# Loại bỏ trùng lặp dựa trên cột 'Name'
df_drop_dup = df_concat.drop_duplicates(subset=['Name'])
print("\nSau khi loại bỏ trùng lặp tên:\n", df_drop_dup)

# Sắp xếp theo cột 'Age'
df_sorted = df_concat.sort_values(by='Age')
print("\nSau khi sắp xếp theo Age:\n", df_sorted)

# Đổi tên cột
df_rename = df_concat.rename(columns={'Name': 'Surname', 'Age': 'Age_pupil'})
print("\nSau khi đổi tên cột:\n", df_rename)

# Xóa bớt cột dữ liệu
df_dropped = df_concat.drop(columns=['Age'])
print("\nSau khi xóa cột Age:\n", df_dropped)
```

Gom nhóm dữ liệu DataFrame (Groupby)

```python
# Hàm Groupby kết hợp các hàm tính toán cơ bản (Mô phỏng lại ảnh mẫu Iris/Fish từ bài học)
# Tạo nhanh một DataFrame mẫu để chạy trực tiếp:
data_sample = {
    'Species': ['Bream', 'Bream', 'Roach', 'Roach', 'Pike'],
    'Weight': [242.0, 290.0, 150.0, 120.0, 500.0],
    'Length': [25.4, 26.3, 22.0, 21.0, 41.0]
}
df_fish = pd.DataFrame(data_sample)

# Thực hiện gom nhóm theo Species và tính trung bình (mean)
print("Giá trị trung bình theo loài:")
print(df_fish.groupby(['Species']).mean())

print("\nĐếm số lượng phần tử theo loài:")
print(df_fish.groupby(['Species']).count())
```

```python
import pandas as pd
import numpy as np

# Tạo dữ liệu mô phỏng chính xác theo hình ảnh thực hành trong bài
data_fish = {
    'Species': ['Bream', 'Bream', 'Bream', 'Bream', 'Bream', 'Roach', 'Roach', 'Pike', 'Pike'],
    'Weight': [242.0, 290.0, 340.0, 363.0, 430.0, 150.0, 120.0, 500.0, 716.0],
    'Length': [25.4, 26.3, 26.5, 29.0, 29.0, 22.0, 21.0, 41.0, 42.7],
    'Height': [11.5200, 12.4440, 12.3778, 12.7300, 12.4440, 8.9520, 8.1457, 10.7388, 11.5200],
    'Width': [4.0200, 3.8417, 4.6961, 4.4555, 5.1340, 3.2928, 3.1234, 4.4555, 5.5891]
}

df_csv = pd.DataFrame(data_fish)
print("--- DataFrame gốc ban đầu ---")
print(df_csv)
```

```python
# Gom nhóm theo Species và tính trung bình (mean) của từng loài
print("--- Giá trị trung bình của từng nhóm Loài (Species) ---")
df_mean = df_csv.groupby(['Species']).mean()
print(df_mean)

# Đếm số lượng mẫu dữ liệu thuộc mỗi loài dựa theo cột Weight
print("--- Số lượng mẫu dữ liệu của mỗi loài ---")
df_count = df_csv.groupby(['Species'])['Weight'].count()
print(df_count)

# 1. Tìm giá trị nhỏ nhất (Min) của từng loài
print("--- Giá trị nhỏ nhất (Min) theo từng loài ---")
print(df_csv.groupby(['Species']).min())

# 2. Tìm giá trị lớn nhất (Max) của từng loài
print("\n--- Giá trị lớn nhất (Max) theo từng loài ---")
print(df_csv.groupby(['Species']).max())

# 3. Tính độ lệch chuẩn (Standard Deviation)
print("\n--- Độ lệch chuẩn (Std) theo từng loài ---")
print(df_csv.groupby(['Species']).std())
```
